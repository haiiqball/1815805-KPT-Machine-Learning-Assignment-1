```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image 
from IPython.display import Image
%matplotlib inline
```

```python
coin = cv2.imread('coin.jpg')
coin_gray = cv2.cvtColor(coin, cv2.COLOR_BGR2GRAY)
#coin_gray = cv2.cvtColor(coin, cv2.COLOR_BGR2RGB)
plt.imshow(coin_gray, cmap = 'gray')
plt.show()
```

```python
coin_blur = cv2.GaussianBlur(coin_gray, (11,11), 0) 
plt.imshow(coin_blur, cmap = 'gray')
plt.show()
```

```python
coin_canny = cv2.Canny(coin_blur, 5,150,3)
plt.imshow(coin_canny,cmap = 'gray')
plt.show()
```

```python
coin_dilated = cv2.dilate(coin_canny,(1,1), iterations = 20)
plt.imshow(coin_dilated,cmap = 'gray')
```

```python
(countour, hierarchy) = cv2.findContours(coin_dilated.copy(),cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_NONE)
coin_rgb = cv2.cvtColor(coin,cv2.COLOR_BGR2RGB)
cv2.drawContours(coin_rgb,countour, -1, (0,255,0),2)

plt.imshow(coin_rgb)
```
